```python
c,o,l,x=input().split()
```

```python
c=int(c)
o=int(o)
l=int(l)
x=int(x)
```

```python
C=c//4
O=o//8
L=l//2
X=x
```

```python
if C<=O and C<=L and C<=X:
  N=C
if O<=C and O<=L and O<=X:
  N=O
if L<=C and L<=O and L<=X:
  N=L
if X<=C and X<=O and X<=L:
  N=X
```

```python
S=N*1259
```

```python
h=S//3600
m=(S-(h*3600))//60
s=S%60
```

```python
print('%d h %d m %d s'%(h,m,s))
```

```python
N1,N2,N3,N4,N5,N6=input().split()
```

```python
N1=int(N1)
N2=int(N2)
N3=int(N3)
N4=int(N4)
N5=int(N5)
N6=int(N6)
```

```python
n=N1+N2+N3+N4+N5+N6
```

```python
if n<40:
  print("B-")
elif 40<=n<60:
  print("B")
elif 60<=n<80:
  print("B+")
elif 80<=n<100:
  print("A-")
elif 100<=n<150:
  print("A")
elif 150<=n<180:
  print("A+")
elif 180<=n<200:
  print("S-")
elif 200<=n<250:
  print("S")
elif 250<=n:
  print("S+")
```
