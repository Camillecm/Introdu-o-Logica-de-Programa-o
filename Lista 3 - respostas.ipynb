```python
N=int(N)
X=int(X)
Xpmin=int(Xpmin)
```

```python
for i in range(N):
  XP,Q=input().split()
  XP=int(XP)
  Q=int(Q)
  if XP>=Xpmin:
    print(XP+X, Q+1)
  else:
    print(XP,Q)
```
