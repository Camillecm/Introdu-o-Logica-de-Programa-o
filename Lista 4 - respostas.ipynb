```python
N=int(input())
X=input().split()
M=int(input())
X=list(map(int,X))
```

```python
vida=M
i=0
while i<N:
  if X[i]>1:
    vida=vida-X[i]
    if vida<1:
      break
  elif X[i]==0:
      i=i+1
      continue
  else:
      vida=M
  i=i+1

if vida<1:
  print("You Died")
else:
  print("Yes, you can")
```

F - excedeu o tempo limite

```python
N=int(input())
X=input().split()
X=list(map(int,X))
```

```python
L=[]
while sum(X)!=N*100000000:
  for i in range(N):
    if X[i]==min(X):
      L.append(X[i])
      X[i]=100000000
```

```python
H=[]
for i in range(len(L)):
  if L[i]<100000000:
    H.append(L[i])
print(*H,end=" ")
```
