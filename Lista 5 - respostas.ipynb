```python
P=int(input())
```

```python
L=[]
for i in range(10):
  H=input().split()
  H=list(map(int,H))
  L.extend(H)
  H=[]
```

```python
achou=False
i=0
while i<len(L):
  if L[i]==P:
    achou=True
    break
  i+=1
if achou:
  print("sim")
else:
  print("nao")
```

```python
linhas,colunas=input().split()
linhas=int(linhas)
colunas=int(colunas)
```

```python
matriz_A = [[int(i) for i in input().split()] for i in range(linhas)]
```

```python
matriz_B = [[int(i) for i in input().split()] for i in range(linhas)]
```

```python
for i in range(linhas):
  for j in range(colunas):
    matriz_A[i][j]=matriz_A[i][j]-matriz_B[i][j]

for i in range(linhas):
  matriz=list(map(int,matriz_A[i]))
  print(*matriz)
```

```python
N=int(input())
```

```python
P=[[int(i) for i in input().split()] for i in range(N)]
```

```python
X,Y=input().split()
X=int(X)
Y=int(Y)
```

```python
ron=0
for i in range(N):
  ron=ron+P[i][Y]

harry=sum(P[X])

if X<=Y:
  harry=harry-P[X][Y]
else:
  ron=ron-P[X][Y]

print("Harry %d" % (harry))
print("Ron %d" % (ron))
```

```python
N,M=input().split()
N=int(N)
M=int(M)
```

```python
L=[[int(i) for i in input().split()] for i in range(N)]
```

```python
soma=0
i=0
while i<N:
  if i%2==0:
    for j in range(0,M,1):
      soma=soma+L[i][j]
      if soma<0:
        soma=0
  else:
    for j in range(M-1,-1,-1):
      soma=soma+L[i][j]
      if soma<0:
        soma=0
  i=i+1
print(soma)
```
