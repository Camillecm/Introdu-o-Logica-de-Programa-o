```python
N=int(input())
```

```python
L=[]
for i in range(N):
  grau,dem=input().split()
  L.append({"grau": int(grau),
            "dem": int(dem)})
```

```python
for inicio in range(1, N):
   i = inicio
   while i >= 1 and (L[i]["dem"] > L[i-1]["dem"] or (L[i]["dem"] == L[i-1]["dem"] and L[i]["grau"] > L[i-1]["grau"])):
       L[i], L[i-1] = L[i-1], L[i]
       i -= 1

for i in L:
   print(i["grau"], i["dem"])
```

```python
N=int(input())
catalogo=[]
for i in range(N):
  n=input().split()
  catalogo.append(n)
```

```python
Q=int(input())
especies=[]
for i in range(Q):
  n=input().split()
  especies.append(n)
```

```python
esq = 0
dir = N - 1

for i in range(Q):
  while esq <= dir:
    meio = (esq + dir) // 2
    if catalogo[meio] == especies[i]:
      break
    elif especies[i] < catalogo[meio]:
      dir = meio - 1
    else:
      esq = meio + 1
      
  if catalogo[meio] == especies[i]:
    especies[i].append("vive!")
  else:
    especies[i].append("foi extinto :(")
  esq = 0
  dir = N - 1
```

```python
for i in range(Q):
  print(*especies[i])
```

```python
n = len(L)
esq = 0
dir = n - 1
for i in range(Q):
  while esq < dir:
     meio = esq + (dir - esq) // 2
     if H[i] >= L[meio]:
         esq = meio + 1
     else:
        dir = meio
  if L[esq] >= H[i]:
    print(esq)
  else:
    print(len(L))
  esq = 0
  dir = n - 1
```
