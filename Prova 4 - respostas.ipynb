```python
N=int(input())
U=input().split()
V=input().split()
D=input().split()
```

```python
U=list(map(int,U))
V=list(map(int,V))
D=list(map(int,D))
L=[]
for i in range(N):
  a=U[i]+V[i]
  L.append(a)
  a=0
```

```python
if L==D:
  print("OK")
else:
  print("NOPE :(")
```

```python
N=int(input())
n=input().split()
n=list(map(int,n))
Q=int(input())
```

```python
i=0
vskin=0
while vskin<=Q:
  vskin=vskin+n[i]
  i=i+1
vskin=vskin-n[i-1]
print(vskin//100)
```

```python
N=int(input())
n=input().split()
n=list(map(int,n))
Q=int(input())
```

```python
soma =0
for i in range(len(n)):
  soma=soma+n[i]
  if soma>=Q:
    break

print((soma-n[i])//100)
```
