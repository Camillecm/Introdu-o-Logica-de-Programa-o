```python
tabuleiro=[[int(i) for i in input().split()] for i in range(8)]
```

```python
A,B=input().split()
A=int(A)
B=int(B)
```

```python
h=0
for i in range(B+1,8):
  if tabuleiro[A][i]==1:
    break
  if tabuleiro[A][i]==2:
    h=h+1
    break

for i in range(B-1,-1,-1):
  if tabuleiro[A][i]==1:
    break
  if tabuleiro[A][i]==2:
    h=h+1
    break
```

```python
for i in range(A+1,8):
  if tabuleiro[i][B]==1:
    break
  if tabuleiro[i][B]==2:
    h=h+1
    break

for i in range(A-1,-1,-1):
  if tabuleiro[i][B]==1:
    break
  if tabuleiro[i][B]==2:
    h=h+1
    break
print(h)
```
