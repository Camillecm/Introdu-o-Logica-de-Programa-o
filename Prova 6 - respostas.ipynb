```python
N,C=input().split()
N=int(N)
C=int(C)
```

```python
L=[]
for i in range(N):
  arvores,altura=input().split()
  L.append({"arvores": arvores,
            "altura": int(altura)})
```

```python
for inicio in range(1, N):
   i = inicio
   while i >= 1 and (L[i]["altura"] > L[i-1]["altura"] or (L[i]["altura"] == L[i-1]["altura"] and L[i]["arvores"] > L[i-1]["arvores"])):
       L[i], L[i-1] = L[i-1], L[i]
       i -= 1

while len(L)!=C:
  L.pop(-1)

for i in L:
   print(i["arvores"])
```

```python

N=int(input())
```

```python
total=[]
for i in range(N):
  n=input().split()
  total.extend(n)
```

```python
M=int(input())
```

```python
cartas=[]
for i in range(M):
  n=input().split()
  cartas.extend(n)
```

```python
esq = 0
dir = N - 1
K=0
for i in range(M):
  while esq <= dir:
    meio = (esq + dir) // 2
    if total[meio] == cartas[i]:
      break
    elif cartas[i] < total[meio]:
      dir = meio - 1
    else:
      esq = meio + 1
      
  if total[meio] == cartas[i]:
    K=K+1
  esq = 0
  dir = N - 1

if K==M:
  print("Jaden pode se tornar o Rei dos Jogos")
else:
  print("Jaden nao pode se tornar o Rei dos Jogos")
```
